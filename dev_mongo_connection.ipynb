```python
lengths = []
for doc in mongo_collection.aggregate([
    {

        '$unwind': {
            'path': '$description'
        }
    }, {
        '$project': {
            'description': 1
        }
    }
]):
    lengths.append(len(doc['description']))

lengths[:10]
```
