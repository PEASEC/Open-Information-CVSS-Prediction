```python
from mongo_handler import MongoHandler
import requests
from bs4 import BeautifulSoup, Tag
```

```python
mongo_handler = MongoHandler()
intel_references = mongo_handler.get_reference_list_for_url("www.intel.com")
print(f"Found {len(intel_references)} references")
intel_references
```

```python
url = intel_references[0]['url']
cve_id = intel_references[0]['_id']
print(cve_id, url)
website = requests.get(url)
website
```

```python
intel_soup = BeautifulSoup(website.content, 'html.parser')
intel_soup.prettify()
```

```python
summary_header = intel_soup.find(name='h2', text=lambda s: "Summary" in s)
print(summary_header)
```

```python
summary_block = summary_header.findNext('p')
full_text = summary_block.text
```

```python
details_header = intel_soup.find(name='h2', text=lambda s: 'Vulnerability Details' in s)
details_header
```

```python
id_block: Tag
for block in details_header.next_siblings:
    if not block == "\n":
        if block.name == "h2":
            break
        if block.name == "p" and cve_id in block.text:
            details_text= block.find_next("p").text
            full_text += details_text
            break

full_text
```
