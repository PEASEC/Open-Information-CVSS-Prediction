```python
import random

from selenium.webdriver.common.by import By
from selenium.webdriver.support.relative_locator import locate_with
from trafilatura import fetch_url, extract

from mongo_handler import setup_mongo_connection, get_reference_list_for_url
from selenium_scraper import setup_driver
```

```python
mongo_collection = setup_mongo_connection()
qualcomm_references = get_reference_list_for_url("www.qualcomm.com", mongo_collection)
print(f"Found {len(qualcomm_references)} matching reference urls") # [{'_id': 'CVE-2021-XYZ', 'url': 'https://wpscan.com/wtfxyz42'}]
qualcomm_references
```

```python
filtered_references = []
for e in qualcomm_references:
    url: str = e['url']
    if url.endswith('bulletin'):
        filtered_references.append(e)
```

```python
driver = setup_driver()
```

```python
ref = random.choice(filtered_references)
url = ref['url']
#url = 'https://www.qualcomm.com/company/product-security/bulletins/july-2021-bulletin'
#url = 'http://www.ibm.com/support/docview.wss?uid=swg2C1000134'
cve_id = ref['_id']
#cve_id = 'CVE-2021-1889'
#cve_id = 'CVE-2016-0202'
driver.get(url)
print(f'{cve_id}: {url}')
```

```python
headers = driver.find_elements(By.TAG_NAME, 'h4')
headers
```

```python
for i, e in enumerate(headers):
    if e.text == cve_id:
        correct_header = e

correct_header
```

```python
tds = driver.find_elements(locate_with(By.TAG_NAME, 'td').below(correct_header))
tds
```

```python
td = tds[2]
text = td.text
text
```

```python
driver.quit()
```

```python
trafilatura_downloaded = fetch_url(url)
if trafilatura_downloaded is not None:
    trafilatura_result = extract(trafilatura_downloaded)
    print(trafilatura_result)
```
