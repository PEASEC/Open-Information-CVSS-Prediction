```python
from mongo_handler import MongoHandler
import requests
from bs4 import BeautifulSoup, Tag
```

```python
mongo_handler = MongoHandler()
talos_references = mongo_handler.get_reference_list_for_url("talosintelligence.com")
print(f"Found {len(talos_references)} references")
talos_references
```

```python
url = talos_references[888]['url']
cve_id = talos_references[1000]['_id']
url = "https://talosintelligence.com/vulnerability_reports/TALOS-2018-0690"
cve_id = "CVE-2018-4019"
print(url)
website = requests.get(url)
website
```

```python
soup = BeautifulSoup(website.content, 'html.parser')
soup.prettify()
```

```python
summary = soup.find(id='summary')
summary_parent_div = summary.parent
summary_parent_div
```

```python
summary_ps = summary_parent_div.find("p")
summary_ps.get_text()
```

```python
details: Tag = soup.find(id='details')
details
```

```python
full_text: str = ""
h4s = summary_parent_div.findAll("h4")
stop_on_next = False
if len(h4s):
    for sib in details.next_siblings:
        if not sib == "\n":
            if sib.text == "Timeline":
                break
            if sib.name == "h4" and stop_on_next:
                break
            if sib.name == "h4" and cve_id in sib.text:
                stop_on_next = True
            if sib.name == "p":
                full_text += sib.text
else:
    for sib in details.next_siblings:
        if not sib == "\n":
            if sib.text == "Timeline":
                break
            if sib.name == "p":
                print(sib)
                full_text += sib.text
print("----------")
print(full_text)
```

```python
full_text = summary_ps.get_text() + full_text
full_text
```
